#### Use grouping instead of pivot

```python
week_day_groups = bike.groupby("weekday")
```

```python
week_day_groups.groups
```

```python
week_day_groups.groups.keys()
```

```python
week_day_groups.get_group(5)
```

```python
week_data = week_day_groups[["registered", "casual"]].mean()
week_data.sort_index(axis=0)
```

```python
week_data["registered"].plot.bar()
plt.xticks(ticks=[0, 1, 2, 3, 4, 5, 6], labels = ["Sunday", "Monday", "Tuesday", "Wednesday",
                                                  "Thursday", "Friday", "Saturday"], rotation = 90)
plt.title("Bike Sharing Pattern")
plt.xlabel("Days")
plt.ylabel("Counts")
plt.show()
```

```python
week_data["registered"].plot.barh()

plt.yticks(ticks=[0, 1, 2, 3, 4, 5, 6], labels = ["Sunday", "Monday", "Tuesday", "Wednesday",
                                                  "Thursday", "Friday", "Saturday"], rotation = 0)
plt.title("Bike Sharing Pattern")
plt.xlabel("Counts")
plt.ylabel("Days")
plt.show()
```
