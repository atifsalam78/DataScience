```python
avg_deaths = pd.read_csv("avg_deaths.csv") # 1st January to 15 December 2020
avg_deaths
```

```python
avg_deaths["Num_days"] = 0
```

```python
avg_deaths["Year"] = 2020
```

```python
avg_deaths.columns
```

```python
avg_deaths = avg_deaths[['Year','Month', 'Num_days', 'New_deaths']]
```

```python
avg_deaths
```

```python
import calendar
_, num_days = calendar.monthrange(2020,1)
```

```python
num_days
```

```python
import calendar
def num_of_days(year, month):
    _, num_days = calendar.monthrange(year, month)
    return num_days
```

```python
num_of_days(2020,2)
```

```python
avg_deaths.iloc[0]["Month"]
```

```python
days = avg_deaths[["Year", "Month"]].apply(lambda row: num_of_days(row["Year"], row["Month"]), axis=1)
```

```python
days
```

```python
avg_deaths.Num_days = days
```

```python
avg_deaths.loc[11, "Num_days"]//=2
```

```python
avg_deaths
```

```python
avg_deaths["Monthly_deaths"] = avg_deaths.Num_days.mul(avg_deaths.New_deaths)
```

```python
avg_deaths
```

```python
avg_deaths.rename({"New_deaths": "Avg_deaths"}, axis=1, inplace=True)
```

```python
avg_deaths["Cum_deaths"] = avg_deaths.Monthly_deaths.cumsum()
```

```python
avg_deaths
```

```python
# Jan - March 1 - 3
# March - Jun 3 - 6
# Jun - Sep 6 - 9
# Sep - Dec 9 - 12
```

```python
# avg_deaths.Month[:3]
# avg_deaths.Month[2:6]
# avg_deaths.Month[5:9]
avg_deaths.Month[8:]
```

```python
cum_cases = [avg_deaths.Cum_deaths.loc[2], avg_deaths.Cum_deaths.loc[5],
             avg_deaths.Cum_deaths[8], avg_deaths.Cum_deaths[11]]
cum_cases
proportion = [round(i/cum_cases[-1],2) for i in cum_cases]
proportion
xmaxs = [0.5+i*(0.8-0.5) for i in proportion]
xmaxs
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=4, ncols=1, figsize=(6, 8))

axes = [ax1, ax2, ax3, ax4]

for ax in axes:
    ax.plot(avg_deaths["Month"], avg_deaths["Avg_deaths"], color="#b00b1e", alpha=0.1)
    ax.tick_params(left=False, bottom=False)
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    for sp_val in ax.spines.values():
        sp_val.set_visible(False)
ax1.plot(avg_deaths.Month[:3], avg_deaths.Avg_deaths[:3], color="#b00b1e", linewidth=2.5)
ax1.text(x=0.5, y=-80, s="0", alpha=0.5, color="grey")
ax1.text(x=1.1,y=-350, s="Jan-Mar", rotation=2)
ax1.text(0.5, 3500, "The Virus Kill 900 people everyday", size = 14, weight = "bold")
ax1.text(0.5, 3150, "Average Number of Daily Deaths", size = 12,)


ax2.plot(avg_deaths.Month[2:6], avg_deaths.Avg_deaths[2:6], color="#b00b1e", linewidth=2.5)
ax2.text(x=3.7, y=2000, s="1844", alpha=0.5, color="grey")
ax2.text(x=3.8,y=500, s="Mar-Jun", rotation=2)

ax3.plot(avg_deaths.Month[5:9], avg_deaths.Avg_deaths[5:9], color="#b00b1e", linewidth=2.5)
ax3.text(x=7.6, y=1200, s="1020", alpha=0.5, color="grey")

ax4.plot(avg_deaths.Month[8:], avg_deaths.Avg_deaths[8:], color="#b00b1e", linewidth=2.5)
ax4.text(x=11.5, y=2400, s="2247", alpha=0.5, color="grey")

for ax, death, xmax in zip(axes, cum_cases, xmaxs):
    ax.axhline(y=2000, xmin=0.5, xmax=0.8, linewidth=6, alpha=0.1, color="#b00b1e")
    ax.axhline(y=2000, xmin=0.5, xmax=xmax, linewidth=6, color="#b00b1e")
    ax.text(x=7.5, y=2200, s=f"{death:,}", weight="bold")

plt.show()
    
```

```python
url = "https://archive.ics.uci.edu/static/public/186/wine+quality.zip"
```

```python
import requests
response = requests.get(url)
response
```

```python
with open("wine.zip", "wb") as f:
    f.write(response.content)
```

```python
import zipfile
with zipfile.ZipFile("wine.zip") as zipped:
    zipped.extractall("wine_data")
```

```python
red_wine = pd.read_csv("wine_data/winequality-red.csv", sep=";")
white_wine = pd.read_csv("wine_data/winequality-white.csv", delimiter=";")
```
