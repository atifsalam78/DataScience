# Design Principles
* Familarity Principle: Easy for reader
* Maximizing Dark Ink Ratio

```python
fig, ax2 = plt.subplots(figsize=(4.5, 6))

ax2.barh(deaths.Country_Other, deaths.Total_Deaths, height = 0.50, color="#800000")
ax2.tick_params(left= False, bottom=False) # To remove ticks from both x an y-axis
for sp_val in ax2.spines.values(): # To remove all four spines
    sp_val.set_visible(False)
# ax.set_xticks([0, 150000, 300000])
ax2.set_xticks(np.linspace(0, 300000, 3)) # To set x ticks values
ax2.set_xticklabels(["0", "150,000", "less than 300,000"])
ax2.xaxis.tick_top() # to shift ticks from bottom to top
ax2.tick_params(top=False) # to remove ticks
ax2.tick_params(axis= "x", colors="grey") # make only x-axis less visible 

ax2.set_yticklabels([])
coutries = deaths["Country_Other"]

for ind, val in enumerate(countries):
    ax2.text(x=-80000, y=ind-0.17, s=val)

# plt.title("Covid Deaths") # Non-Customizeable
# ax.text(x = -80000, y = 22, s = "Worldwide Covid Death Toll")
ax2.text(-80000, 23.5, "Worldwide Covid Deaths", size=17, weight ="bold")
ax2.text(-80000, 22.5, "Top 20 countries by death toll - December 2020", size=12 )
ax2.axvline(150000, ymin=0.045, ymax=1, color="grey", alpha=0.5)

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(11, 6))

ax1.barh(deaths.Country_Other, deaths.Total_Deaths)

ax2.barh(deaths.Country_Other, deaths.Total_Deaths, height = 0.50, color="#800000")
ax2.tick_params(left= False, bottom=False) # To remove ticks from both x an y-axis
for sp_val in ax2.spines.values(): # To remove all four spines
    sp_val.set_visible(False)
# ax.set_xticks([0, 150000, 300000])
ax2.set_xticks(np.linspace(0, 300000, 3)) # To set x ticks values
ax2.set_xticklabels(["0", "150,000", "less than 300,000"])
ax2.xaxis.tick_top() # to shift ticks from bottom to top
ax2.tick_params(top=False) # to remove ticks
ax2.tick_params(axis= "x", colors="grey") # make only x-axis less visible 

ax2.set_yticklabels([])
coutries = deaths["Country_Other"]

for ind, val in enumerate(countries):
    ax2.text(x=-80000, y=ind-0.17, s=val)

# plt.title("Covid Deaths") # Non-Customizeable
# ax.text(x = -80000, y = 22, s = "Worldwide Covid Death Toll")
ax2.text(-80000, 23.5, "Worldwide Covid Deaths", size=17, weight ="bold")
ax2.text(-80000, 22.5, "Top 20 countries by death toll - December 2020", size=12 )
ax2.axvline(150000, ymin=0.045, ymax=1, color="grey", alpha=0.5)

plt.subplots_adjust(wspace=0.5)
plt.show()
```
