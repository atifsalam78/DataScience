```python
# Ploting using pandas
work_day_mean["casual"].plot.bar()
plt.show()
```

```python
work_day_mean["casual"].plot.bar()
plt.xticks(rotation = 0)
plt.show()
```

```python
work_day_mean["casual"].plot.bar()
plt.xticks(ticks=[0,1], labels=["Holiday", "Working Day"], rotation = 0)
plt.show()
```
